```python
car = cv2.imread('car.jpg')
car.shape
```

```python
plt.imshow(car)
plt.show()
```

```python
carRGB = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
plt.imshow(carRGB)
plt.show()
```

```python
plt.imshow(car[...,2],cmap='gray')
plt.show()
```

```python
carGray = car[...,0] * 0.114 + car[...,1] * 0.587 + car[...,2] * 0.299
carGray = carGray.astype(np.uint8)
plt.imshow(carGray,cmap='gray')
plt.show()
```

```python
carGray = cv2.cvtColor(car,cv2.COLOR_BGR2GRAY)
plt.imshow(carGray,cmap='gray')
plt.show()
```
